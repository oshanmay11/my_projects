```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import kagglehub
path = kagglehub.dataset_download("arjitdsce/churn-modelling")
```

```python
df = pd.read_csv(path + "/Churn_Modelling.csv")
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
geography = pd.get_dummies(df["Geography"],drop_first=True)
gender = pd.get_dummies(df["Gender"],drop_first=True)
```

```python
# Reconstruct the feature set 'x' as a DataFrame from the original 'df'
# Select all columns from 'df' except 'RowNumber', 'CustomerId', 'Surname', 'Geography', 'Gender', and the target 'Exited'
columns_to_keep = [col for col in df.columns if col not in ["RowNumber", "CustomerId", "Surname", "Geography", "Gender", "Exited"]]
x_base = df[columns_to_keep]

# Concatenate with the one-hot encoded geography and gender DataFrames
x = pd.concat([x_base, geography, gender], axis=1)

# The original 'Geography' and 'Gender' columns (from df) have now been replaced
# by their one-hot encoded versions, and other irrelevant columns are dropped.
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LeakyReLU,PReLU,ELU
```

```python
#initializing the ANN
classifier = Sequential()
```

```python
#adding the input and the 1st hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu',input_dim = 11))
```

```python
#addingg the second layer
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))
```

```python
#adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
```

```python
#compiling the ANN
classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#fitting the ANN to the training set
model_history=classifier.fit(x_train, y_train,validation_split=0.33, batch_size = 5, epochs = 100)
```

```python
#predicting the test set results
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
